```python
import pandas as pd
from datetime import datetime
```

```python
dbutils.widgets.get("input") 
myinput_uri = getArgument("input")

dbutils.widgets.get("output")
myoutput_uri = getArgument("output")

dbutils.widgets.get("input_blob_config")
myinput_blob_config = getArgument("input_blob_config")

dbutils.widgets.get("output_blob_config")
myoutput_blob_config = getArgument("output_blob_config")

dbutils.widgets.get("input_blob_secretname") 
myinput_blob_secretname = getArgument("input_blob_secretname")

dbutils.widgets.get("output_blob_secretname") 
myoutput_blob_secretname = getArgument("output_blob_secretname")
```

```python
try:
    dbutils.fs.mount(
      source = myinput_uri,
      mount_point = "/mnt/input",
      extra_configs = {myinput_blob_config:dbutils.secrets.get(scope = "amlscope", key = myinput_blob_secretname)})
except Exception as e:
    print(e)
    pass

try:
    dbutils.fs.mount(
      source = myoutput_uri,
      mount_point = "/mnt/output",
      extra_configs = {myoutput_blob_config:dbutils.secrets.get(scope = "amlscope", key = myoutput_blob_secretname)})
except Exception as e:
    print(e)
    pass
```

```python
df = pd.read_csv('/dbfs/mnt/input/input.csv')
display(df)
```

```python
timestamp = datetime.strftime(datetime.now(), '%Y-%m-%d %H:%M:%S')
new_df = pd.DataFrame([{'A':timestamp}])
final_df = pd.concat([df, new_df])
final_df.to_csv('/dbfs/mnt/output/output.csv', index=False)
display(final_df)
```
